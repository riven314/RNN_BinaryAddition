### Summary
Try different hyper-parameters and compare the performance by analysis of the metrics plot. With different experiments done on hyper-parameters, I try to draw conclusion on the choice of hyper-parameters (see bottom for the final conclusions)

### Remark
Unlike original code, I use minibatch(with randomised shuffling) instead of looping through each data point one by one. This analysis takes the choice of minibatch as default because it could help speed up convergence and at the same time leverage the computational efficiency from vectorisation. It serves as an optimisation on convergence speed and computational speed.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from data_util import *
from model_util import *
from metrics_util import *
```

```python
FILENAME = 'data/' + FILENAME
binary_dim = 8    
batch_size = 64
n_epoch = 40
n_mb = 40
num4train = batch_size * n_mb
num4test = 100
is_print_metrics = False
```

#### Training Variables

```python
### TRAINING SESSION
tf.reset_default_graph()
train_X, train_Y = init_placeholder()
parameters = init_parameters(hidden_dim = 4,
                                          init_method = 'xavier',
                                          rand_seed = 1)
rnn_cell = create_BasicRNN(hidden_dim = [4, 4, 4],
                                        activation = 'tanh')
outputs, state = forward_timestep(train_X, rnn_cell, batch_size = batch_size)
train_h = forward_output(train_X, outputs, parameters)
train_cost = compute_cost(train_h, train_Y)
train_step = backpropagate_optimise(train_cost,
                                                     lr = 0.01,
                                                     optimiser = 'adam')
```

#### Test Variables

```python
### TESTING SESSION
test_X, test_Y = init_placeholder()
test_output, _ = forward_timestep(test_X, rnn_cell, batch_size = num4test)
test_h = forward_output(test_X, test_output, parameters)
test_cost = compute_cost(test_h, test_Y)
```

#### Load Data

```python
# pull data
data_dict = txt_2_dict(FILENAME)
x, y = dict_2_nparr(data_dict, 
                           is_flip = True)
train_x, train_y, test_x, test_y = partition_data(x, y, 
                                                  train_idx = 0, num4train = num4train, 
                                                  test_idx = 4000, num4test = num4test)
```

#### Run Training

```python
# start training
sess = tf.Session()
sess.run(tf.global_variables_initializer())

train_cost_list, test_cost_list = [], []
train_acc_list, test_acc_list = [], []
for i in range(n_epoch):
    train_x, train_y = shuffle(train_x, train_y, rand_seed = i)
    for j in range(n_mb):
        mb_x, mb_y = get_minibatches(train_x, train_y, j, batch_size)
        # training set
        _, train_bs_probs, train_bs_cost = sess.run([train_step, train_h, train_cost], 
                                                                    feed_dict = {train_X: mb_x, train_Y: mb_y})
        train_cost_list.append(train_bs_cost)
        train_preds = get_prediction(train_bs_probs)
        train_acc = get_accuracy(mb_y, train_preds)
        train_acc_list.append(train_acc)
        # test set
        test_all_probs, test_all_cost = sess.run([test_h, test_cost] , feed_dict = {test_X: test_x, test_Y: test_y})
        test_cost_list.append(test_all_cost)
        test_all_preds = get_prediction(test_all_probs)
        test_acc = get_accuracy(test_y, test_all_preds)
        test_acc_list.append(test_acc)
        # printing
        if j % 10 == 0:
            print('Epoch: %d Example: %d is running...' % (i,j))
        if is_print_metrics:
            print('test_cost = {} | test_acc = {}'.format(test_all_cost, test_acc))
            print('train_cost = {} | train_acc = {}'.format(train_bs_cost, train_acc))
        
sess.close()
```

#### is_flip = True v.s. is_flip = False
[DIFFERENCE] is_flip <br>
[COMMON] lr = 0.01, init = xavier, optimiser = adam, bs_size = 64, n_mb = 40, n_epoch = 40, hidden act = elu, hidden_dim = [4] <br><br>
When data x, y are flipped upside down, the test loss and accuracy improved significantly particularly in early stage. It is because the timestep data are read from top to bottom, it assumes that the data at timestep at upper level depends on the data at timestep at lower level. In binary arithmetic rule, the number on left depends on the number on right. Hence, it makes to flip the data upside down. Also, the dependence is localised, meaning that a number only depends on the number right next to it. Therefore, we don't have a concern about long-distant dependence.

#### hidden_dim = [4] v.s. [4, 4] v.s. [4, 4, 4]
[DIFFERENCE] hidden_dim <br>
[COMMON] lr = 0.01, init = xavier, optimiser = adam, bs_size = 64, n_mb = 40, n_epoch = 40, hidden act = elu, is_flip = True <br><br>

When hidden layers change from 1 layer [4] to 2 layer [4, 4], accuracy increases significantly (an approx increase of 10%). It is because the model is more complex so it can learn more complex pattern. The improvement is sharp at the early training. Also, there is no sign of overfitting from the loss curve. It shows that we do not need regularization technique such as dropout at the moment. 
The improvement margin for accuracy is small when hidden layers change from 2 layer [4, 4] to 3 layers [4, 4, 4]. So improvement margin for cost, it is halved.

```python
# lr = 0.01, init = xavier, optimiser = adam, bs_size = 64, n_mb = 40, n_epoch = 40, hidden act = elu, is_flip = True, hidden_dim = [4, 4. 4]
get_metrics_plot(train_cost_list, test_cost_list, train_acc_list, test_acc_list)
print('test accuracy: {}'.format(test_acc_list[-1]))
print('test loss: {}'.format(test_cost_list[-1]),3)
```

#### Conclusion Drawn from The Analysis
1. Don't use reLU/ eLU/ leaky ReLU for hidden activation. it leads to gradient explosion
2. tanh as activation function in hidden layer gives the best performance 
3. Flipping x, y upside down significantly improves the training performance
4. Adam with learning rate = 0.01 is good enough
5. Choice of optimiser doesn't give much difference
6. Weight initialisation method doesn't give much difference
7. Using 2 hidden layers with [4, 4] shows significant improvement over 1 hidden layer with [4]
8. Hidden layers changing from [4, 4] to [4, 4, 4] doesn't give much improvement margin
